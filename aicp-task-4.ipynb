```python
import math
class Hexagon:
    def __init__(self, side_length):
        self.side_length = side_length

    def calcArea(self):
        return 1.5 * 1.732 * self.side_length ** 2

    def calcPeri(self):
        return 6 * self.side_length

    def calcAngleSum(self):
        return 6 * 120

    def display(self):
        print(f"Hexagon - Area: {self.calcArea()}, Perimeter: {self.calcPeri()}, Sum of Angles: {self.calcAngleSum()}")

# Square class
class Square:
    def __init__(self, side_length):
        self.side_length = side_length

    def calcArea(self):
        return self.side_length ** 2

    def calcPeri(self):
        return 4 * self.side_length

    def display(self):
        print(f"Square - Area: {self.calcArea()}, Perimeter: {self.calcPeri()}")

# Main menu
while True:
    cnic_last_digit = int(input("Enter the last digit of your CNIC: "))
    hexagon = Hexagon(cnic_last_digit)
    square = Square(cnic_last_digit + 1)

    print("Menu:")
    print("1. Display area, perimeter, and sum of angles of hexagon")
    print("2. Display area and perimeter of square")
    print("Any other key to exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        hexagon.display()
    elif choice == '2':
        square.display()
    else:
        print("Exiting the program.")
        break
```
